## ROI(Region of Interest)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imgBGR = cv2.imread("Commons.jpeg", cv2.IMREAD_COLOR)
img = imgBGR[:,:,::-1]

ROI_x, ROI_y = eval( input( "Enter (x, y) for ROI: " ) ) # 100, 100
ROI_nr, ROI_nc = eval( input( "Enter (rows, columns) for ROI: " ) ) # 200, 300

ROI = img[ ROI_x : ROI_x + ROI_nr, ROI_y : ROI_y + ROI_nc ]

plt.figure(figsize=(12, 10))
plt.subplot(2,2,1)   
plt.title('original image') 
plt.imshow(img,cmap=plt.cm.gray)
plt.subplot(2,2,2)   
plt.title('ROI image') 
plt.imshow(ROI,cmap=plt.cm.gray)
```

## Use opencv: puttext to embed text into an image 
Reference: https://docs.opencv.org/2.4/modules/core/doc/drawing_functions.html#cv2.putText

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 定義數位影像(全黑)
img = np.zeros( [ 400, 500, 3 ], dtype = 'uint8' )
type(img)
# 置入文字
text = "Hello OpenCV"
fontFace = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText( img, text, ( 10, 50 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_PLAIN
cv2.putText( img, text, ( 10, 90 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_DUPLEX
cv2.putText( img, text, ( 10, 130 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_COMPLEX
cv2.putText( img, text, ( 10, 170 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_TRIPLEX
cv2.putText( img, text, ( 10, 210 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_COMPLEX_SMALL
cv2.putText( img, text, ( 10, 250 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
cv2.putText( img, text, ( 10, 290 ), fontFace, 1.0, ( 255, 255, 255 ) )
fontFace = cv2.FONT_HERSHEY_SCRIPT_COMPLEX
cv2.putText( img, text, ( 10, 330 ), fontFace, 1.0, ( 255, 255, 255 ) )


plt.figure(figsize=(12, 10))
plt.subplot(2,2,1)   
plt.title('original image') 
plt.imshow(img,cmap=plt.cm.gray)
plt.subplot(2,2,2)   
plt.title('ROI image') 
plt.imshow(img,cmap=plt.cm.gray)
```

    <td>image size:100x512, gray levels from left to right ranging between:0~255</td>
    </tr></table>    

```python
# Homework 1
import numpy as np
import cv2

img = np.tile(np.linspace(0, 255, 512), (100, 1))
# numpy.linspace(start, stop, num, ...)
# Returns num evenly spaced samples, calculated over the interval [start, stop].
# numpy.tile(A, reps)
# Construct an array by repeating A the number of times given by reps.

plt.figure(figsize=(12, 10))
plt.title('original image') 
plt.imshow(img, cmap=plt.cm.gray)
```

```python
import numpy as np
import cv2

img = np.zeros([100, 512], dtype="uint8")
nr, nc = img.shape[:2]

for x in range( nr ): # the height
    for y in range( 0, nc-2, 2 ): # the width and every two columns are the same color
        img[x][y+1] = img[x][y] # the first block(y) and the second block(y+1) are the same color in this row
        img[x][y+2] = img[x][y+1]+1 # the third block(y+2) is lighter than others in this row
    
plt.figure(figsize=(12, 10))
plt.title('original image') 
plt.imshow(img, cmap=plt.cm.gray)
```

## <font color="Red"> Homework 2: Please create the following two images by Python</font><p>

    <td>image size:512x512, the image only with gray levels of two colors 0 and 255 as binary image</td>
</tr></table>    

```python
# Homework 2
black = np.zeros((64, 64, 3), np.uint8)
white = black.copy()
white.fill(255)

rowA = np.concatenate((black, white), axis=1)
for x in range(2):
    rowA = np.concatenate((rowA, rowA), axis=1)

rowB = np.concatenate((white, black), axis=1)
for x in range(2):
    rowB = np.concatenate((rowB, rowB), axis=1)

binary = np.concatenate((rowA, rowB), axis=0)
for x in range(2):
    binary = np.concatenate((binary, binary), axis=0)

plt.figure(figsize=(12, 10))
plt.title('original image') 
plt.imshow(binary, cmap=plt.cm.gray)
```

```python
image = np.zeros([512, 512, 3], dtype="uint8")
nr, nc = image.shape[:2]

for x in range( nr ):
    for y in range( nc ):
        if (x // 64 + y // 64) % 2 == 0:
            image[x][y] = 0
        else:
            image[x][y] = 255

plt.figure(figsize=(12, 10))
plt.title('original image') 
plt.imshow(image, cmap=plt.cm.gray)
```

## <font color="Red"> Homework 3: Please create the following two images by Python</font><p>
Please use the folloeing bit extraction method to acquire the bit plane image: <p>
val = 63 <p>
bit_plane = 0b00000001 <p>
val & bit_plane = 1 <p>
<img src="Chapter01_h3.png" width="800px">    

```python
# Homework 3
Lenna = cv2.imread("Lenna.bmp", 0)
r, c = Lenna.shape

# 產生 bit-plane 對應的 mask
x = np.zeros((r,c), dtype = np.uint8)
# print("x.ndim = ", x.ndim) # x.ndim =  2

# 建立 1 個空白的 bit plane
r = np.zeros((r,c), dtype = np.uint8)

# 建立 bit plane 7
# r[:,:] 代表著被分割出來的 bit plane
# r2 用作將 r[:,:] bit plane 大於 0 的元素轉變為白色(用於顯示圖片)，會比較清晰。
x = 2**7
r[:,:] = cv2.bitwise_and(Lenna, x)
mask = r[:,:] > 0
r2 = np.copy(r)
r2[mask] = 255    

plt.figure(figsize=(12, 10))
plt.subplot(2,2,1)   
plt.title('source image') 
plt.imshow(Lenna, cmap=plt.cm.gray)
plt.subplot(2,2,2)   
plt.title('image of bit plane 7') 
plt.imshow(r2[:,:], cmap=plt.cm.gray)
```
